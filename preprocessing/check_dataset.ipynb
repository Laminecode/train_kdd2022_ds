```python
#check dataset.py
import os

def check_dataset(images_dir, labels_dir):
    images = [f for f in os.listdir(images_dir)
              if f.endswith(('.jpg', '.png', '.jpeg'))]
    labels = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

    images_ids = set(os.path.splitext(f)[0] for f in images)
    labels_ids = set(os.path.splitext(f)[0] for f in labels)

    missing_labels = images_ids - labels_ids
    missing_images = labels_ids - images_ids

    print(len(images), "images trouvées.")
    print(len(labels), "labels trouvés.")
    print(f"[INFO] Images sans labels : {len(missing_labels)}")
    print(f"[INFO] Labels sans images : {len(missing_images)}")

    return missing_labels, missing_images
```

```python
images_dir = "../../dataset/raw/train/images"
labels_dir = "../../dataset/raw/train/labels"

missing_labels, missing_images = check_dataset(images_dir, labels_dir)

print("====== DATASET INTEGRITY CHECK ======")
print(f"Images sans labels : {len(missing_labels)}")
print(f"Labels sans images : {len(missing_images)}")

if len(missing_labels) > 0:
    print("Exemples images sans labels :", list(missing_labels)[:5])

if len(missing_images) > 0:
    print("Exemples labels sans images :", list(missing_images)[:5])
```
